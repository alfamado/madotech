a) compute the following: (cA), (dB) and (c+d)A
### Solution
```
(cA) =     2[2 -1]
            [0 3]
          = [4 -2]
            [0 6]

(dB) = 3[1 4]
        [-2 5]

    = [3 12]
      [-6 15]

(c+d)A = (2+3)[2 -1]
              [0 3]

      = 5[2 -1]
         [0 3]

      = [10, -5]
        [0 15]
```

b) Using the results gotten, verify the scalar multiplication properties
# Solution
```
1) Associativity of scalar multiplication: (c.d)A = c.(dA)<br>
(c.d)A = (2*3)[2 -1]
              [0 3]

        = 6[2 -1]
            [0 3]

        = [12 -6]
          [0 18]

(dA)= 3[2 -1]
       [0 3]

= [6 -3]<br>
  [0 9]

c. (dA)=   2[6 -3]
            [0 9]

= [12 -6]
  [0 18]

from the above solution we can see that the associativity of scalar multiplication is correct. our (c.d)A is equal to our c. (dA)

2) Distributivity over matrix addition: c(A+B) = cA + cB
### Solution
(A+B) =
[2 -1] + [1 4]
[0 3]  + [-2 5]

= [3 3]
  [-2 8]

c(A + B) =
2[3 3]
 [-2 8]

= [6 6]
  [-4 16]

(cA) =     2[2 -1]
            [0 3]
          = [4 -2]
            [0 6]

(cB) = 2[1 4]
        [-2 5]
      = [2 8]
        [-4 10]

cA + cB =
=[4 -2] + [2 8]
[0 6] + [-4 10]

= [6 6]
[-4 16]

from the above solution we can see that the distributivity over matrix addition is correct. our c(A+B) is equal to cA + cB

3) Distributivity over scalar addition:
(c+d)A =
(2+3)[2 -1]
     [0 3]
  = 5[2 -1]
     [0 3]
  = [10, -5]
    [0 15]

(cA) =     2[2 -1]
            [0 3]
          = [4 -2]
            [0 6]

(dA) =     3[2 -1]
            [0 3]
          = [6 -3]
            [0 9]

cA + dA =
[4 -2] + [6 -3]
[0 6] + [0 9]

= [10 -5]
[0 15]

from the above solution we can see that the distributivity over scalar addition is correct. our (c+d)A is equal to cA + dA
```
